```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 데이터프레임 df4가 이미 존재한다고 가정합니다.

# 모든 범주형 변수 라벨 인코딩
label_encoder = LabelEncoder()

categorical_columns = ['substitute0', 'substitute1', 'substitute2', 'substitute3', 'substitute4',
                       'use0', 'Chemical Class', 'Therapeutic Class', 'Action Class', 'sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']

for col in categorical_columns:
    df4[col] = label_encoder.fit_transform(df4[col].astype(str))

# 독립 변수와 종속 변수 설정
features = ['substitute0', 'substitute1', 'substitute2', 'substitute3', 'substitute4',
            'use0', 'Chemical Class','Therapeutic Class', 'Action Class']
X = df4[features]
side_effects = ['sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, df4[side_effects], test_size=0.2, random_state=42)

# 표준화
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 변수 중요도를 저장할 리스트 생성
decision_tree_importances = []
random_forest_importances = []

# 모든 sideEffect에 대해 모델 학습 및 중요도 계산
for effect in side_effects:
    y_train_effect = y_train[effect]
    y_test_effect = y_test[effect]

    # 의사결정나무 회귀
    tree_reg = DecisionTreeRegressor()
    tree_reg.fit(X_train, y_train_effect)
    y_pred_tree = tree_reg.predict(X_test)
    decision_tree_importances.append(tree_reg.feature_importances_)

    mse_tree = mean_squared_error(y_test_effect, y_pred_tree)
    r2_tree = r2_score(y_test_effect, y_pred_tree)
    print(f"\nDecision Tree Regression for {effect}:")
    print("MSE:", mse_tree)
    print("R2 Score:", r2_tree)

    # 랜덤 포레스트 회귀
    forest_reg = RandomForestRegressor(n_estimators=100)
    forest_reg.fit(X_train, y_train_effect)
    y_pred_forest = forest_reg.predict(X_test)
    random_forest_importances.append(forest_reg.feature_importances_)

    mse_forest = mean_squared_error(y_test_effect, y_pred_forest)
    r2_forest = r2_score(y_test_effect, y_pred_forest)
    print(f"\nRandom Forest Regression for {effect}:")
    print("MSE:", mse_forest)
    print("R2 Score:", r2_forest)

# 각 모델의 중요도 평균 계산
mean_decision_tree_importances = np.mean(decision_tree_importances, axis=0)
mean_random_forest_importances = np.mean(random_forest_importances, axis=0)

# 중요도를 데이터프레임으로 변환하여 출력
importances_decision_tree = pd.DataFrame(mean_decision_tree_importances, index=features, columns=['Importance']).sort_values('Importance', ascending=False)
print("\nDecision Tree Feature Importance:")
print(importances_decision_tree)

importances_random_forest = pd.DataFrame(mean_random_forest_importances, index=features, columns=['Importance']).sort_values('Importance', ascending=False)
print("\nRandom Forest Feature Importance:")
print(importances_random_forest)

# 중요도 시각화 - 색상 변경
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple']  # 원하는 색상 리스트

plt.figure(figsize=(12, 8))
plt.bar(importances_decision_tree.index, importances_decision_tree['Importance'], color=colors[:len(importances_decision_tree)])
plt.title('Mean Decision Tree Feature Importance')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.show()

plt.figure(figsize=(12, 8))
plt.bar(importances_random_forest.index, importances_random_forest['Importance'], color=colors[:len(importances_random_forest)])
plt.title('Mean Random Forest Feature Importance')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 데이터프레임 df4가 이미 존재한다고 가정합니다.

# 모든 범주형 변수 라벨 인코딩
label_encoder = LabelEncoder()

# 범주형 변수들의 리스트
categorical_columns = ['substitute0', 'substitute1', 'substitute2', 'substitute3', 'substitute4',
                       'use0', 'Chemical Class', 'Therapeutic Class', 'Action Class', 'sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']

# 각 범주형 변수에 대해 라벨 인코딩을 수행
for col in categorical_columns:
    df4[col] = label_encoder.fit_transform(df4[col].astype(str))

# 독립 변수와 종속 변수 설정
features = ['substitute0', 'substitute1', 'substitute2', 'substitute3', 'substitute4',
            'use0', 'Chemical Class','Therapeutic Class', 'Action Class']

# 종속 변수들의 리스트
side_effects = ['sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']

# 데이터 분할 (훈련 세트와 테스트 세트로 분할)
X_train, X_test, y_train, y_test = train_test_split(df4[features], df4[side_effects], test_size=0.2, random_state=42)

# 표준화 객체 생성
scaler = StandardScaler()

# 훈련 데이터 표준화 (훈련 세트를 기준으로 맞춤)
X_train = scaler.fit_transform(X_train)

# 테스트 데이터 표준화 (훈련 세트의 맞춤을 기준으로 변환)
X_test = scaler.transform(X_test)

# 변수 중요도를 저장할 리스트 생성
decision_tree_importances = []
random_forest_importances = []

# 모든 sideEffect에 대해 모델 학습 및 중요도 계산
for effect in side_effects:
    # 현재 sideEffect에 해당하는 종속 변수 설정
    y_train_effect = y_train[effect]
    y_test_effect = y_test[effect]

    # 의사결정나무 회귀 모델 생성 및 학습
    tree_reg = DecisionTreeRegressor(random_state=42)
    tree_reg.fit(X_train, y_train_effect)

    # 예측 수행
    y_pred_tree = tree_reg.predict(X_test)

    # 의사결정나무 회귀 모델의 변수 중요도 저장
    decision_tree_importances.append(tree_reg.feature_importances_)

    # 성능 평가 (MSE, R^2)
    mse_tree = mean_squared_error(y_test_effect, y_pred_tree)
    r2_tree = r2_score(y_test_effect, y_pred_tree)
    print(f"\nDecision Tree Regression for {effect}:")
    print("MSE:", mse_tree)
    print("R2 Score:", r2_tree)

    # 랜덤 포레스트 회귀 모델 생성 및 학습
    forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
    forest_reg.fit(X_train, y_train_effect)

    # 예측 수행
    y_pred_forest = forest_reg.predict(X_test)

    # 랜덤 포레스트 회귀 모델의 변수 중요도 저장
    random_forest_importances.append(forest_reg.feature_importances_)

    # 성능 평가 (MSE, R^2)
    mse_forest = mean_squared_error(y_test_effect, y_pred_forest)
    r2_forest = r2_score(y_test_effect, y_pred_forest)
    print(f"\nRandom Forest Regression for {effect}:")
    print("MSE:", mse_forest)
    print("R2 Score:", r2_forest)

# 각 모델의 변수 중요도 평균 계산
mean_decision_tree_importances = np.mean(decision_tree_importances, axis=0)
mean_random_forest_importances = np.mean(random_forest_importances, axis=0)

# 변수 중요도를 데이터프레임으로 변환하여 출력
importances_decision_tree = pd.DataFrame(mean_decision_tree_importances, index=features, columns=['Importance']).sort_values('Importance', ascending=False)
print("\nDecision Tree Feature Importance:")
print(importances_decision_tree)

importances_random_forest = pd.DataFrame(mean_random_forest_importances, index=features, columns=['Importance']).sort_values('Importance', ascending=False)
print("\nRandom Forest Feature Importance:")
print(importances_random_forest)

# 중요도 시각화 - 색상 변경
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple']  # 원하는 색상 리스트

# 의사결정나무 변수 중요도 그래프
plt.figure(figsize=(12, 8))
plt.bar(importances_decision_tree.index, importances_decision_tree['Importance'], color=colors[:len(importances_decision_tree)])
plt.title('Mean Decision Tree Feature Importance')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.show()

# 랜덤 포레스트 변수 중요도 그래프
plt.figure(figsize=(12, 8))
plt.bar(importances_random_forest.index, importances_random_forest['Importance'], color=colors[:len(importances_random_forest)])
plt.title('Mean Random Forest Feature Importance')
plt.xticks(rotation=90)
plt.ylabel('Importance')
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 데이터셋 로드
file_path = '/content/medicine_dataset.csv'
df = pd.read_csv(file_path)

# 필요한 범주형 변수만 라벨 인코딩
label_encoder = LabelEncoder()

# Chemical Class, sideEffect0, sideEffect1 인코딩
df['Chemical Class'] = label_encoder.fit_transform(df['Chemical Class'].astype(str))
chemical_class_labels = label_encoder.classes_

df['sideEffect0'] = label_encoder.fit_transform(df['sideEffect0'].astype(str))
side_effect0_labels = label_encoder.classes_

df['sideEffect1'] = label_encoder.fit_transform(df['sideEffect1'].astype(str))
side_effect1_labels = label_encoder.classes_

# Chemical Class의 고유 값 확인
unique_chemical_classes = {i: val for i, val in enumerate(chemical_class_labels)}

# sideEffect0의 고유 값 확인
unique_side_effect0 = {i: val for i, val in enumerate(side_effect0_labels)}

# sideEffect1의 고유 값 확인
unique_side_effect1 = {i: val for i, val in enumerate(side_effect1_labels)}

# 디코딩된 Chemical Class 결과
decoded_chemical_classes = {
    27: unique_chemical_classes.get(27, "Unknown"),
    71: unique_chemical_classes.get(71, "Unknown"),
    33: unique_chemical_classes.get(33, "Unknown")
}

# 디코딩된 sideEffect 결과
decoded_side_effects = {
    16: unique_side_effect0.get(16, "Unknown"),
    6: unique_side_effect0.get(6, "Unknown"),
    35: unique_side_effect0.get(35, "Unknown"),
    7: unique_side_effect1.get(7, "Unknown"),
    15: unique_side_effect1.get(15, "Unknown"),
    36: unique_side_effect1.get(36, "Unknown")
}

# 결과 출력
print(f"Chemical Class 27에서 가장 빈도가 높은 SideEffect0 값: 16 [{decoded_side_effects[16]}]")
print(f"Chemical Class 71에서 가장 빈도가 높은 SideEffect0 값: 6 [{decoded_side_effects[6]}]")
print(f"Chemical Class 33에서 가장 빈도가 높은 SideEffect0 값: 35 [{decoded_side_effects[35]}]\n")

print(f"Chemical Class 27에서 가장 빈도가 높은 SideEffect1 값: 7 [{decoded_side_effects[7]}]")
print(f"Chemical Class 71에서 가장 빈도가 높은 SideEffect1 값: 15 [{decoded_side_effects[15]}]")
print(f"Chemical Class 33에서 가장 빈도가 높은 SideEffect1 값: 36 [{decoded_side_effects[36]}]\n")

print(f"Chemical Class 27: [{decoded_chemical_classes[27]}]")
print(f"Chemical Class 71: [{decoded_chemical_classes[71]}]")
print(f"Chemical Class 33: [{decoded_chemical_classes[33]}]")
```

```python
import pandas as pd

# 데이터 불러오기
file_path = '/content/medicine_dataset.csv'
df = pd.read_csv(file_path, low_memory=False)

# 필요한 열 선택
columns_to_use = ['Chemical Class', 'Action Class'] + [f'sideEffect{i}' for i in range(10)]
df = df[columns_to_use]

# 결측값 비율 계산
missing_vals = df.isnull().sum() / len(df)

# 45% 이상 결측값이 있는 열 찾기
missing_more_45 = missing_vals[missing_vals > 0.45]

# 결측값 비율이 45% 이상인 열 삭제
df1 = df.drop(list(missing_more_45.index), axis='columns')

# 결측값을 'unknown'으로 대체
df1.fillna('unknown', inplace=True)

# 중복 값 제거
df1 = df1.drop_duplicates()

# 모든 데이터 소문자로 변환 및 공백 제거
df1 = df1.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)

# 결과 확인
print(df1.head())
print(df1.shape)
```

**Correlation** **matrix** **Heat** **Map** **Result**
결측값이 적은 Side Effects(0-4) <-> Chemical Class <-> Action Class 의 관계를 나타내는 Heat Map

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 불러오기
file_path = '/content/medicine_dataset.csv'
df = pd.read_csv(file_path)

# 필요한 범주형 변수만 라벨 인코딩
label_encoder = LabelEncoder()

# 라벨 인코딩할 변수
categorical_columns = ['Chemical Class', 'Action Class',
                       'sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']

for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col].astype(str))

# 히트맵
plt.figure(figsize=(15, 10))
heatmap_data = df[['Chemical Class', 'Action Class', 'sideEffect0', 'sideEffect1', 'sideEffect2', 'sideEffect3', 'sideEffect4']]
corr_matrix = heatmap_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Chemical Class, Action Class, and Side Effects')
plt.show()
```
